```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error
import lightgbm as lgb
import pprint
```

<h2>Data load</h2>

```python
DATA_PATH_TRAIN = r'C:\Users\user\Downloads\EmEl\Kaggle\30-Days-of-ML-Competition\data\train.csv'
DATA_PATH_TEST = r'C:\Users\user\Downloads\EmEl\Kaggle\30-Days-of-ML-Competition\data\test.csv'

x = pd.read_csv(r'C:\Users\user\Downloads\EmEl\Kaggle\30-Days-of-ML-Competition\data\x.csv')
X_train = pd.read_csv(r'C:\Users\user\Downloads\EmEl\Kaggle\30-Days-of-ML-Competition\data\X_train.csv')
X_valid = pd.read_csv(r'C:\Users\user\Downloads\EmEl\Kaggle\30-Days-of-ML-Competition\data\X_valid.csv')
y_train = pd.read_csv(r'C:\Users\user\Downloads\EmEl\Kaggle\30-Days-of-ML-Competition\data\y_train.csv')
y_valid = pd.read_csv(r'C:\Users\user\Downloads\EmEl\Kaggle\30-Days-of-ML-Competition\data\y_valid.csv')

cat_cols = [col for col in x.columns if 'cat' in col]
num_cols = [col for col in x.columns if 'cont' in col]
```

```python
# feature scaling
sc = StandardScaler()
X_train[num_cols] = sc.fit_transform(X_train[num_cols])
X_valid[num_cols] = sc.transform(X_valid[num_cols])
```

```python
#Set the minimum error arbitrarily large
min = 99999999999999999999999 
count = 0 #Used for keeping track of the iteration number
#How many runs to perform using randomly selected hyperparameters
iterations = 50
for i in range(iterations):
    print('iteration number', count)
    count += 1 #increment count
    # try:
    d_train = lgb.Dataset(X_train, label=y_train) #Load in data
    params = {} #initialize parameters
    params['learning_rate'] = np.random.uniform(0, 0.2)
    params['boosting_type'] = np.random.choice(['gbdt', 'dart', 'goss'])
    params['objective'] = 'regression'
    params['metric'] = 'mae'
    params['sub_feature'] = np.random.uniform(0, 1)
    params['num_leaves'] = np.random.randint(20, 300)
    params['min_data'] = np.random.randint(10, 100)
    params['max_depth'] = np.random.randint(5, 200)
    iterations = np.random.randint(10, 10000)
    print(params, iterations)
#Train using selected parameters
clf = lgb.train(params, d_train, iterations)
y_pred = clf.predict(X_valid) #Create predictions on test set
mae = mean_absolute_error(y_pred, y_valid)
print('MAE:', mae)
if mae < min:
    min = mae
    pp = params 
    # except: #in case something goes wrong
    #     print('failed with')
    #     print(params)
print("*" * 50)
print('Minimum is: ', min)
print('Used params', pp)
```
